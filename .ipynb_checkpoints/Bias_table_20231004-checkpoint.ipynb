```python
data_dir = r'F:\Research\Safegraph_bias\20230916'
save_dir = r'F:\Research\Safegraph_bias\figures_20230917'
```

```python
columns = ['Female', 'Male',
        '15–17', '18–24', '25–34', '35–44', '45–54', '55–64', '>65',
       'White', 'Black', 'Hispanic', 'Asian', 'NoSchool', 'HighSchool',
       'NoCollege', 'Bachelor', 'Graduate', '<20K', '20K–30K', '30K–50K',
       '50K–100K', '>100K']
```

```python
urban_rural_county_xls = os.path.join(data_dir, 'ruralurbancodes2013.xls')
urban_rural_county_df = pd.read_excel(urban_rural_county_xls, dtype={'FIPS':str})
# add the urban column
urban_rural_county_df['Urban'] = urban_rural_county_df['Description'].str[:5].replace({'Metro': 'Urban', 'Nonme': 'Rural'})
 
urban_rural_county_df
```

```python
years = [2018, 2019, 2020, 2021, 2022]
df_new = pd.DataFrame()

# level = 'state'
level = 'county'

for year in years[:]:
    print(year)
    # region_name = r'deep_south'
    csv_file = os.path.join(data_dir, f'monthly_bias_{level}_{year}.csv')
    df = pd.read_csv(csv_file)
    df['variable'] = pd.Categorical(df['variable'], categories=columns, ordered=True)  
    if level == 'state':
        
        df = df.groupby('variable', as_index=True, observed=True).agg(bias_median=('bias', 'median'),
                                                                  bias_max=('bias', 'max'),
                                                                  bias_min=('bias', 'min'),
                                                                 )
        # df_new['variable'] =  df['variable']

    if level == 'county':
        df['FIPS'] = df['county_fips'].astype(str).str.zfill(5)
        df = df.merge(urban_rural_county_df[['FIPS', 'Urban']], on='FIPS')
        df = df.groupby(['variable'], as_index=True, observed=True).agg(bias_median=('bias', 'median'),
        # df = df.groupby(['variable', 'Urban'], as_index=True, observed=True).agg(bias_median=('bias', 'median'),
                                                                  bias_max=('bias', 'max'),
                                                                  bias_min=('bias', 'min'),
                                                                 )
        # df_new[r'Urban/Rural'] = df.iloc[:, 0]
        
 
    df_new[f'{year}'] = df[f'bias_median'].apply('{:.3f}'.format) + ' [' + df[f'bias_min'].apply('{:.3f}'.format).astype(str) + ', ' +  df[f'bias_max'].apply('{:.3f}'.format).astype(str)+ ']'
    # break
    # bias_df
# df_new.set_index('variable', inplace=True)
    
# df_new = df_new.loc[columns]

print("level:", level)
df_new.to_csv(os.path.join(save_dir, f'bias_{level}_2018-2022_table_no_urban.csv'),  index=True)    
df_new
```
