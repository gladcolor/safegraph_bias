## merge other files

```python
import os
import pandas as pd
from natsort import natsorted


monthly_patterns_path = r'L:\SafeGraph_monthly_patterns_2018-2022'
monthly_patterns_new_path = r'H:\SafeGraph_2022monthly\to_cluster\merged_files'
os.makedirs(monthly_patterns_new_path, exist_ok=True)

def get_all_files(data_dir, exts=['.csv'], verbose=True):
    all_files = []
    for root_dir, folders, files in os.walk(data_dir):
        # print(root_dir)
        for f in files:
            for ext in exts:
                ext_len = len(ext)
                if ext == f[-ext_len:]:
                    full_name = os.path.join(root_dir, f)
                    all_files.append(full_name)
                    
    if verbose:
        print("Found files:", len(all_files))
    return sorted(all_files)


# csv_files.pop(9)  # Error when downlaoding: 'L:\\SafeGraph_monthly_patterns_2018-2022\\2022\\09\\06\\SAFEGRAPH\\MP\\home_panel_summary.csv',  
# csv_files.pop(11)   # identical to 11-01:  'L:\\SafeGraph_monthly_patterns_2018-2022\\2022\\11\\22\\SAFEGRAPH\\MP\\home_panel_summary.csv',


# YEAR = 2018

years = [2018, 2019, 2020, 2021, 2022]
# years = [2022]

basename_list = ['home_panel_summary.csv', 'normalization_stats.csv', 'visit_panel_summary.csv']

monthly_files = []
for year in years:
    
    data_dir = fr'L:\SafeGraph_monthly_patterns_2018-2022\{year}'
    
    csv_files = get_all_files(data_dir, exts=['.csv'])
    
    csv_files = natsorted(csv_files, reverse=False)

    print("Year:", year)
 
    for f in csv_files:
        sep = '\\'  # path separate for windows 
        first_day = f.split(sep)[-4]
        # print(first_day, f)
        basename = os.path.basename(f)
        if not basename in basename_list:
            continue
        if first_day != '01':
            # print("Not first day data, skipped:", f)
            continue
        if "SS" in f:  # skip the spend
            # print("Not first day data, skipped:", f)
            continue
#         if not 'core_poi-geometry-patterns-part' in f:
#             # print("Not the Patterns file, skipped:", f)
#             continue
            
        print(f)
        monthly_files.append(f)
        # df = pd.read_csv(f)
        # df_list.append(df)
        
len(monthly_files)
```

```python
# merge files

data_dict = {basename:[] for basename in basename_list}
for bname in basename_list:
    # print(bname)
    data_dict[bname] = []
    for f in tqdm(monthly_files):
        basename = os.path.basename(f)
        data_dict[basename].append(f)
        
for file_type, f_list in data_dict.items():
    
    df_list = []
    for f in f_list:
        df = pd.read_csv(f)
        df_list.append(df)
    
    all_df = pd.concat(df_list)
    
    new_name = os.path.join(monthly_patterns_new_path, file_type)
    all_df.to_csv(new_name, index=False)
    print("Saved file at:", new_name)
```
